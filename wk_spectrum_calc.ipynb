```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from wk_spectra import wk_analysis as wk
```

```python
diri = ''
fili = 'tropics_imerg_v07_6_hourly_mean_1.0deg_2000_2023.nc'

f = xr.open_dataset(diri+fili)
f
```

```python
lati = -15
latf = 15

diro = ''
# filo = 'precip.wkSpectra.20S.20N.nc'

data = f.precipitation.sel(lat=slice(lati, latf)).transpose('time','lat','lon')#.fillna(0)
```

```python
spd = 4
nDayWin = 96
nDaySkip = 65
maxFreq = 0.9
max_wn = 25
```

```python
spectra_obj = wk.wk_analysis(name='wk_analysis')
```

```python
spectra_obj.import_array(array_in=data.values, varname='precipitation')
```

```python
spectra = spectra_obj.wheeler_kiladis_spectra(spd=spd, nDayWin=nDayWin,
                                              nDaySkip=nDaySkip, max_freq=maxFreq,
                                              max_wn=max_wn)
```
